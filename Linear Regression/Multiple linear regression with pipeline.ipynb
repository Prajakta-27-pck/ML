### Seperate numeric and categorical data

```python
cat = []
num = []
for i in x.columns:
    if x[i].dtype == 'object':
        cat.append(i)
    else:
        num.append(i)    
```

#### For Creation of pipeline

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline  =  Pipeline(steps= [('Scaler'  , StandardScaler()) , ('imputer' , SimpleImputer(strategy= 'mean'))])
cat_pipeline  =  Pipeline(steps  = [('encoder' ,  OneHotEncoder(handle_unknown= 'error'))   , ('imputer' , SimpleImputer(strategy= 'most_frequent'))])
```

```python
pre = ColumnTransformer([('num_pipeline' , num_pipeline, num), ('cat_pipeline', cat_pipeline,cat)])
```

```python
pre
```

```python
cols  =  x.columns
cols
```

```python
x1 = pre.fit_transform(x)
```
